#•	Completion Time Frame Categories: How do trial durations distribute across four quantiles from start to completion date?

```python
df['Completion Date']= pd.to_datetime(df['Completion Date'], format ='mixed', errors='coerce')

df['Completion Date']

#df['Start Year'] = df['Start Date'].dt.year
#df['Start Year']
```

```python
df = df.dropna(subset=['Completion Date'])
```

```python
df['Completion Date1'] = df['Completion Date'].dt.year
df['Completion Date1']
```

```python
df['Trial Duration'] = (df['Completion Date'] - df['Start Date']).dt.days
df['Trial Duration']
```

```python
df['Completion Time Frame Categories1']=pd.qcut(df['Trial Duration'],
                                             q=4, precision=0)
```

```python
df['Completion Time Frame Categories1'].unique()
```

```python
bin_labels2 =['Short-term Trials','Medium-term Trials','Long-term Trials','Extended-term Trials']

df['Completion Time Frame Categories']=pd.qcut(df['Trial Duration'],
                                             q=4, precision=0,
                                              labels=bin_labels2)
df['Completion Time Frame Categories']


#(-1.0, 139.0] Short-term Trials
#(139.0, 293.0] Medium-term Trials
#(293.0, 497.0] Long-term Trials
#(497.0, 29119.0] Extended-term Trials
```
